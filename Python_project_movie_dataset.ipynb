```python
#importing libraries 
import pandas as pd 
import seaborn as sns
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure


%matplotlib inline
matplotlib.rcParams['figure.figsize']= (12,8) #Adjusts the configurations of the plots created

#Reading in the data: movie dataset from kaggle
df = pd.read_csv(r"/home/rec1/Desktop/Python/movies.csv")
# top 5 rows
df.head()
```

```python
# Checking for missing data

for col in df.columns:
    #pct_missing1 = df[col].isnull()
    pct_missing = np.mean(df[col].isnull()) # The mean is zero after null returns false which is 0
    # if the mean is greater than zero then there are some missing values
    print('{} - {}%'.format(col,pct_missing))
```

```python
# Data Types for columns
df.dtypes
```

```python
#Replacing NaN with 0
df['budget'] = df['budget'].fillna(0)
# Changing data type from float64 to int64
df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].fillna(0)
df['gross'] = df['gross'].astype('int64')
```

```python
# Extracting year from released column
dfrel2 = df['released'].astype(str).str.split(',',1, expand = True)
df['yearcorrect'] = dfrel2[1].astype(str).str.split('(',1, expand = True)[0]
```

```python
df.head()
```

```python
#Sorting with descending gross values
df.sort_values(by = ['gross'], inplace = False, ascending =False)
```

```python
#To show all rows
#pd.set_option('display.max_rows', None)
#To show distinct values in a column
df['company'].drop_duplicates().sort_values(ascending = False)
```

```python
#To remove duplicates from dataset
df.drop_duplicates()
```

```python
# Scatter plot with budget vs gross 

plt.scatter(x = df['budget'], y = df['gross'], ) #xlabel = 'budget', ylabel = 'gross' not permitted in bracket
# have to define separately
plt.xlabel('Budget for film')
plt.ylabel('Gross earnings')
plt.title('Budget vs gross revenue')
plt.show()
```

```python
#Plot budget vs gross using seaborn
 
sns.regplot(x = 'budget', y = 'gross',  data = df, scatter_kws = {"color":"red"}, line_kws = {"color":"blue"})
```

```python
# Looking at the correlation which only acts on numerical fields
df.corr()
```

```python
 #Visualising correlation matrix
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot = True)
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.title('Correlation Matrix for Numeric Features')
plt.show()
```

```python
# Converting objects in dataframe to numbers
df['name'] = df['name'].astype('object') # the name columns was made up of strings and its data type was category
#so converted it to object before running for loop
df_numerized = df.copy() 

for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] =  df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
df_numerized
```

```python
correlation_matrix = df_numerized.corr()
sns.heatmap(correlation_matrix, annot = True)
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.title('Correlation Matrix for Numeric Features')
plt.show()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
#Looking at pairs with high correlation
threshold = 0.5
high_correlation = sorted_pairs[(sorted_pairs)>threshold]
high_correlation
```
